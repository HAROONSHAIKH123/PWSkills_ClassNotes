# Scrap product link

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime

name = "infinix"
baseurl = 'https://www.flipkart.com'
product_links = []
for i in range(0,1):
    try:
        r = requests.get(f"https://www.flipkart.com/search?q={name}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}")
        soup = BeautifulSoup(r.content,"html.parser")
        span = soup.find_all("div",{"class":"_2kHMtA"})
        for i in range(0,100):
            try:
                link = baseurl + span[i].a['href']
            except:
                continue
            product_links.append(link)
    except:
        continue

        
        
product_datas = []
for i in product_links:
    url = i
    r = requests.get(url)
    try:
        soup = BeautifulSoup(r.content,"html.parser")
    except:
        soup = None
    try:
        product_name = soup.div.h1.text.split("\xa0")[0]
    except:
        product_name = None
    try:
        rating = soup.find("div",{"class":"_3LWZlK"}).text
    except:
        rating = None
    try:
        total_rating = soup.find("span",{"class":"_2_R_DZ"}).text.split("\xa0")[0].split(" ")[0].replace(",","")
    except:
        total_rating = None
    try:
        total_review = soup.find("span",{"class":"_2_R_DZ"}).text.split("\xa0")[::-1][0].split(" ")[0].replace(",","")
    except:
        total_review = None
    try:
        current_price = soup.find("div",{"class":"_30jeq3 _16Jk6d"}).text.replace("₹","").replace(",","")
    except:
        current_price = None
    try:
        warannty = soup.find("div",{"class":"_352bdz"}).text
    except:
        warannty = None
    try:
        highlight = soup.find_all("li",{"class":"_21Ahn-"})[4].text
    except:
        highlight = None
    try:
        seller_rating = soup.find("div",{"class":"_1RLviY"}).div.text
    except:
        seller_rating = None
    try:
        seller_name = soup.find("div",{"class":"_1RLviY"}).span.text[:-3]
    except:
        seller_name = None
    product_data ={"product_name" : product_name,
                "rating":rating,
                "total_rating":total_rating,
                "total_review":total_review,
                "current_price":current_price,
                "warannty":warannty,
                "highlight":highlight,
                "seller_rating":seller_rating,
                "seller_name":seller_name}
    product_datas.append(product_data)
df=pd.DataFrame(product_datas)
```
