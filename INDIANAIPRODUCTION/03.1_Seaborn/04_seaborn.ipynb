# Python seaborn Hist/dist plot 2

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import  norm
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['total_bill'], label='Total Bill')
plt.title('Histogram of total bill')
plt.legend()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], label='Total Bill')
plt.title('Histogram of total bill')
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], bins=bins)
plt.xticks(bins)
plt.title('Histogram of total bill')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--', 'alpha':0.9 })
plt.title('Histogram of total bill')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--', 'alpha':0.9 }, kde_kws={'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9 })
plt.title('Histogram of total bill')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--', 'alpha':0.9 }, kde_kws={'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9 },rug=True, rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9 })
plt.title('Histogram of total bill')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--', 'alpha':0.9 }, kde=False, fit=norm, fit_kws={'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9 } ,rug=True, rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9 })
plt.title('Histogram of total bill')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--', 'alpha':0.9 }, kde=False, fit=norm, fit_kws={'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9 } ,rug=True, rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9 },label='TB')
plt.title('Histogram of total bill', fontsize=20)
plt.xlabel('Total Bill', fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set()
sns.distplot(tips_df['total_bill'], bins=9, label='Total Bill')
sns.distplot(tips_df['tip'], bins=9, label='Tip')
sns.distplot(tips_df['size'], bins=9, label='size')
plt.legend()
```
