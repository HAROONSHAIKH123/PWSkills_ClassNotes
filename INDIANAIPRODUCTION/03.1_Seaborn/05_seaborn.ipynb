# Python seaborn Bar plot

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import  norm
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.barplot(y=tips_df.total_bill, x=tips_df.day)
```

```python
sns.barplot(y=tips_df.total_bill, x=tips_df.day, hue=tips_df.sex)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex')
```

```python
order = ['Sun', 'Thur', 'Fri', 'Sat']
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', order=order)
```

```python
hue_order = ['Female', 'Male']
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', hue_order=hue_order)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', estimator=np.mean)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', estimator=np.max)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', ci=20)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', estimator=np.max, n_boot=2)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', orient='v')
```

```python
sns.barplot(data=tips_df,x='total_bill', y='day', hue='sex')
```

```python
sns.barplot(x='total_bill', y='size', hue='sex', data=tips_df, orient='h')
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', color='g')
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', palette='hot')
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', palette='magma')
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', saturation=0.3)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', errcolor='0.5')
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex',errwidth=12)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', capsize=0.5)
```

```python
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex',dodge=False)
```

```python
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'} 
sns.barplot(data=tips_df,y='total_bill', x='day', hue='sex', **kwargs)
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'} 
sns.barplot(data=tips_df,y='total_bill', x='day', **kwargs)
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'} 
sns.barplot(data=tips_df,y='total_bill', x='day', alpha=0.9, linestyle='-.', linewidth=6, edgecolor='g')
```

```python
ax = sns.barplot(data=tips_df,y='total_bill', x='day', alpha=0.9, linestyle='-.', linewidth=6, edgecolor='g')

ax.set(title = 'Barplot of tips dataframe', xlabel='Days', ylabel='Total Bill')
```

```python
plt.figure(figsize=(16,9))
sns.barplot(data=tips_df,y='total_bill', x='day', alpha=1, linestyle='-.', linewidth=2, edgecolor='k')
plt.title('Bar plot of total bill',fontsize=20)
plt.xlabel('Days',fontsize=15)
plt.ylabel('total Bill',fontsize=15)
plt.show()
# plt.savefig('new')
```
