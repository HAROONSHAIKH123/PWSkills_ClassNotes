# Python seaborn heatmap plot

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import  norm
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(data=arr_2d)
```

```python
globalwarming_df = pd.read_csv('/config/workspace/IndianAiProduction/03.1_Seaborn/data/Who_is_responsible_for_global_warming.csv')
globalwarming_df.head(2)
```

```python
globalwarming_df.columns
```

```python
globalwarming_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1,inplace=True)
```

```python
globalwarming_df.set_index('Country Name', inplace=True)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data=globalwarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, vmin=0, vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap='coolwarm')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap='coolwarm', annot=True)
```

```python
annot_arr = np.array([['a00','a01','a02'],['a10','a11','a12'],['a20','a21','a22'],['a30','a31','a32']])
annot_arr
```

```python
sns.heatmap(arr_2d, annot=annot_arr, fmt='s')
```

# part 2

```python
plt.figure(figsize=(16,9))
annot_kws = {'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.9, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
sns.heatmap(globalwarming_df, cmap='coolwarm', annot=True, annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap='coolwarm', annot=True, linewidths=4)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap='coolwarm', annot=True, linewidths=4, linecolor='k')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cbar=False)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cbar=False, xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }
sns.heatmap(globalwarming_df, cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,9))
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']

sns.heatmap(globalwarming_df, cbar=False, xticklabels=np.arange(0,14), yticklabels=country_code)
```

```python
plt.figure(figsize=(16,9))

ax = sns.heatmap(globalwarming_df)
ax.set(title='Haet map for who is resp for global warming', xlabel='CO2 wmision per year', ylabel='country name')
sns.set(font_scale=.5)
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalwarming_df.corr(), annot=True, linewidths=3)
ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')
plt.title('ABCD', fontsize=25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']], columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(cancer_df.corr(), annot=True, linewidths=3)
ax.tick_params(size=10, color='k', labelsize=10, labelcolor='k')
plt.title('heatap of breast cancer patient', fontsize=25)
plt.show()
```
