# Handling missing values in python pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Fortune_10.csv')
df
```

```python
# assign NaN instead for "not available"
df = pd.read_csv('data/Fortune_10.csv', na_values='not available')
df
```

```python
# assign NaN instead for "not available"
df = pd.read_csv('data/Fortune_10.csv', na_values=['not available','no values'])
df
```

```python
# for specfic column only
df = pd.read_csv('data/Fortune_10.csv', na_values={'Industry':'no values'})
df
```

```python
df = pd.read_csv('data/Fortune_10.csv', na_values={'Industry':'no values', 'Growth':'not available'})
df
```

```python
# To keep default CSV Nan value instaed of Nan
df = pd.read_csv('data/Fortune_10.csv', keep_default_na=False)
df
```

```python
df = pd.read_csv('data/Fortune_10.csv', na_filter=False)
df
```
