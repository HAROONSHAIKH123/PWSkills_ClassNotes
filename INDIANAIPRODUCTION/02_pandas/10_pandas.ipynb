# Pandas fill NA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Fortune_05.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name':'None', 'Growth':0})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(axis=0, method='ffill')
```

```python
df.fillna(axis=1, method='ffill')
```

```python
df.fillna(axis=1, method='bfill')
```

```python
df.fillna(0,limit=1)
```

```python
df.fillna(0,limit=2)
```
