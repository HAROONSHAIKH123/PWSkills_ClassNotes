# Pandas concate()

```python
import pandas as pd
import numpy as np
```

```python
sr1 = pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
sr3 = pd.Series([3,4,5,6,7])
sr3
```

```python
pd.concat([sr1,sr3])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name': ['a','b','c','d'],
             'class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'name': ['e','f','g','h'],
             'class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df1,df2], axis=0,ignore_index=True)
```

# Day-2

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name': ['a','b','c','d'],
             'class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[3,4],
             'name': ['c','d'],
             'class':[7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df1,df2], axis=1,join='inner')
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'name': ['e','f','g','h'],
             'class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2], keys=['df1', 'df2'])
```

```python
pd.concat([df1,df2], keys=['first_df', 'second_df'])
```

```python
pd.concat([df1,df2],axis=1, keys=['first_df', 'second_df'])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name': ['a','b','c','d'],
             'class':[5,6,7,8]})
df1
```
