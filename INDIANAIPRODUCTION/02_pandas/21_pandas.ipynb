# pandas melt()

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/movie_rating.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=['genre'])
```

```python
pd.melt(df, id_vars=['year'])
```

```python
pd.melt(df, id_vars=['year'], value_vars='genre')
```

```python
pd.melt(df, id_vars=['year'], value_vars='rating %')
```

```python
pd.melt(df, id_vars=['year'], value_vars='rating %', var_name='catagory')
```

```python
pd.melt(df, id_vars=['year'], value_vars='rating %', var_name='catagory', value_name='data')
```
