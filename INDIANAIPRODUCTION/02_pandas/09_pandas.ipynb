# Handling missing value in python

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Fortune_05.csv')
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='any')
```

```python
df.dropna(axis=1, how='any')
```

```python
# remove row if row have all nan value
df.dropna(how='all')
```

```python
# min 1 notnull value presernt
df.dropna(thresh=1)
```

```python
# min 4 notnull value presernt
df.dropna(thresh=4)
```

```python
df.dropna(subset=['ID'])
```

```python
df.dropna(subset=['Name'])
```

```python
df.dropna(subset=['Companies'])
```
