# 20 Pandas join()

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'a':[1,2,3],
                   'b':[10,20,30]})
df2 = pd.DataFrame({'c':[4,5,6],
                   'd':[40,50,60]})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1 = pd.DataFrame({'a':[1,2,3],
                   'b':[10,20,30]},
                  index=['a','b','c'])
df2 = pd.DataFrame({'c':[4,5,6],
                   'd':[40,50,60]})
display(df1,df2)
```

```python
# index should be same
df1.join(df2)
```

```python
df1 = pd.DataFrame({'a':[1,2,3],
                   'b':[10,20,30]},
                  index=['a','b','c'])
df2 = pd.DataFrame({'c':[4,5,6],
                   'd':[40,50,60]},
                  index=['a','b','c'])
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1 = pd.DataFrame({'a':[1,2,3],
                   'b':[10,20,30]},
                  index=['a','b','c'])
df2 = pd.DataFrame({'c':[4,5],
                   'd':[40,50]},
                  index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='right')
```

```python
df1.join(df2,how='inner')
```

```python
df1.join(df2,how='outer')
```

```python
df1 = pd.DataFrame({'a':[1,2,3],
                   'b':[10,20,30]},
                  index=['a','b','c'])
df2 = pd.DataFrame({'a':[4,5],
                   'd':[40,50]},
                  index=['a','b'])
display(df1,df2)
```

```python
# column name are same will give error use suffix
df1.join(df2,lsuffix='_1')
```

```python
df1.join(df2,rsuffix='_1')
```
