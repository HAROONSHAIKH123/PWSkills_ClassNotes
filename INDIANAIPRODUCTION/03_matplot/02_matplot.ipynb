# matplot histogram

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,40,(100))
print(ml_students_age)
print(py_students_age)
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='bar', orientation='horizontal', color='m',label='ML students')
plt.title('ML students age histograme')
plt.xlabel('Student age catagory')
plt.ylabel('Nos students age')
plt.legend(loc=4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='bar', orientation='vertical', color='m',label='ML students')
plt.hist(py_students_age,bins, rwidth=0.8, histtype='bar', orientation='vertical', color='y',label='PY students')

plt.title('ML students age histograme')
plt.xlabel('Student age catagory')
plt.ylabel('Nos students age')
plt.legend(loc=4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]


plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8, histtype='bar', orientation='vertical', color=['m','y'],label=['ML students','py students'])
# plt.hist(py_students_age,bins, rwidth=0.8, histtype='bar', orientation='vertical', color='y',label='PY students')

plt.title('ML students age histograme')
plt.xlabel('Student age catagory')
plt.ylabel('Nos students age')
plt.legend(loc=4)
plt.show()
```
